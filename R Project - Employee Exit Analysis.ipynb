
<h3>Predicting whether employee leaves or does not leave</h3>

```r
library(randomForest)
library(e1071)

library(caTools)
library(readr)
library(caret)
```

```r
split <- sample.split(hr$left, SplitRatio = 0.75)
```

```r
train <- subset(hr,split == T)
```

```r
cv <- subset(hr, split == F)
```

```r
table(cv$left)
```

```r
cv
```

### logistic regression

```r
glm.model <- glm(left ~ ., data = train, family = "binomial")
summary(glm.model)
```

```r
glm.predict <- predict(glm.model, cv, type = "response")
table(cv$left, glm.predict > 0.41)
```

```r
glm.accuracy <- (2499+478)/nrow(cv)
paste("Accuracy of glm model is:",glm.accuracy*100, "%", sep = " ")
```

### random forest model

```r
set.seed(55)
rf.model <- randomForest(left ~ ., data = train, ntree = 2000, nodesize = 1)
rf.predict <- predict(rf.model, cv)
confusionMatrix(cv$left, rf.predict)
```

```r
install.packages("Boruta", repos='http://cran.us.r-project.org')
```

```r
library(Boruta)
set.seed(13)
bor.model <- Boruta(left ~ ., data = train, maxRuns=101, doTrace=0)
```

```r
summary(bor.model)
```

```r
boruta.cor.matrix <- attStats(bor.model)
important.features <- names(train)[bor.model$finalDecision!="Rejected"]
```
